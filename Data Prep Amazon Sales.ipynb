### Handling Missing Value

Dikarenakan terdapat missing value pada fitur atau kolom yang akan digunakan untuk kebutuhan analisis, maka akan dilakukan treatment terhadap missing value dengan berlandaskan domain knowledge dengan melakukan pengecekan kembali pada fitur yang terdapat missing value.

```python
### cek pada fitur courier status

amazon['Courier Status'].unique()
```

value dari fitur courier status adalah
- nan
- Shipped
- Cancelled
- Unshiped

dikarenakan data tersebut tidak bersifat atau bertipe numerikal dan value tersebut berisikan mengenai status pengiriman barang dan bersifat krusial, maka missing value dari fitur tersebut tidak akan dihilangkan atau *drop missing value*, karena dari data source kita tidak menemukan informasi yang lebih detail, dikhawatirkan apabila kita melakukan treatment seperti *drop missing value* atau mengganti value tersebut dengan modus dari data tersebut, maka akan mempengaruhi hasil dari analisis yang akan dilakukan pada proses selanjutnya.

```python
### cek pada fitur currency

amazon['currency'].unique()
```

value dari fitur currency adalah
- INR
- nan

Dari peninjauan terhadap value pada fitur currency dan dari domain knowledge, maka missing value berupa *nan* akan digantikan dengan value *INR*. INR sendiri adalah kode mata uang dari negara India, pada data yang sudah di ekstraksipun menjelaskan bahwa data dari amazon sales report ini sendiri menjelaskan bahwa data tersebut data transaksi penjualan yang di kirimkan ke negara India.

```python
### Cek pada fitur Amount

amazon['Amount'].unique()
```

fitur Amount merupakan fitur yang berisi data numerikal, maka akan dilakukan analisis lebih mendalam seperti :
- Harga jenis barang yang serupa
- Dari kota mana dikirimkan
- Harga barang untuk setiap unitnya
- Harga pengiriman untuk setiap unitnya

dari domain knowledge yang dimiliki, disini tidak mendapatkan data yang menjelaskan darimana barang dikirimkan dan berapa harga pengiriman untuk setiap paketnya.

Apabila diisi dengan nilai dari median harga, ditakutkan analisis yang akan kita lakukan nanti tidak tepat sesuai dengan sasaran seperti : harga komoditas ditakutkan meleset sehingga mempengaruhi keputusan yang akan diambil. Sehingga akan dilakukan *drop missing value*

```python
### Cek pada fitur ship-city

amazon['ship-city'].unique()
```

Dari peninjuan terhadap value pada fitur ship-city, fitur tersebut berisikan mengenai nama kota tujuan dari pengiriman barang atau paket, dan untuk jumlah *jumlah missing value* nya hanya berjumlah 33, sehingga akan dilakukan *drop missing value"

### Cek Outliers

```python
amazon.describe() ## deskripsi dari data yang bersifat numerikal
```

```python
q1 = amazon.describe()['Amount']['25%']
q3 = amazon.describe()['Amount']['75%']
iqr = q3-q1
upper_f = q3 + (1.5 * iqr)
lower_f = q1 - (1.5 * iqr)
out_a = amazon[(amazon['Amount'] < lower_f) | (amazon['Amount'] > upper_f)]
out_a
```

```python
import matplotlib.pyplot as plt
```

```python
plt.boxplot(amazon['Amount'])
plt.show()
```

banyak ditemukan outlier pada fitur Amount, untuk nilai tertingginya yaitu 5.600
